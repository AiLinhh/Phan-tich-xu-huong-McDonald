```python
#convert string to dict

import ast
def check_react(react, i):
    try:
        r = ast.literal_eval(post_df_full1['reactions'][i])[react]
    except:
        r = 0
    return r

post_df_full_with_reactions = pd.DataFrame()
for i in range(len(post_df_full1)):


    post_df_full_with_reactions = post_df_full_with_reactions.append([{  
                'post_text': post_df_full1['post_text'][i], 'time': post_df_full1['time'][i],         
                'th√≠ch':check_react('th√≠ch', i),
                'y√™u th√≠ch':check_react('y√™u th√≠ch', i),
                'haha':check_react('haha', i),
                'wow': check_react('wow', i), 
                'bu·ªìn':check_react('bu·ªìn', i),
                'ph·∫´n n·ªô':check_react('ph·∫´n n·ªô', i),
                'th∆∞∆°ng th∆∞∆°ng':check_react('th∆∞∆°ng th∆∞∆°ng', i),
                'shares':post_df_full1['shares'][i],
                'comments':post_df_full1['comments'][i],
                'reaction_count':post_df_full1['reaction_count'][i]
    }])
```
